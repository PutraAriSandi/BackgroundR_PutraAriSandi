## Proses Remove

```python
import matplotlib.pyplot as sandi

input_path = 'apel.jpg'
output_path = 'newfotoapel.png'

# Buka gambar asli
input_image = Image.open(input_path)

# Tampilkan gambar asli sebelum penghapusan latar belakang
sandi.subplot(1, 2, 1)
sandi.imshow(input_image)
sandi.title('Original')

# Hapus latar belakang
output_image = remove(input_image)

# Menaampilkan gambar setelah penghapusan latar belakang
sandi.subplot(1, 2, 2)
sandi.imshow(output_image)
sandi.title(' Removal Background')

# Simpan gambar yang telah dihapus latar belakang
output_image.save(output_path)

# Menampilkan plot
sandi.tight_layout()
sandi.show()
```
